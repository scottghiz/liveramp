```python
df4 = pd.DataFrame().assign(Country=df0['Country'], Current_Population=df0['2022 Population'])
df5 = df4.sort_values('Current_Population', ascending=False)
df_current_pop = df5
df_current_pop.reset_index(drop=True, inplace=True)
#print(df_current_pop)
```

```python
df6 = pd.DataFrame().assign(Country=df0['Country'], Population_1970=df0['1970 Population'])
df7 = df6.sort_values('Population_1970', ascending=False)
df_1970_pop = df7
df_1970_pop.reset_index(drop=True, inplace=True)
#print(df_1970_pop)
```

```python
### merge >> join, concat ###
df_result = pd.merge(df_current_pop, df_1970_pop, on='Country')
#print(df_result)
```

```python
def f(x):
    return (x[1]/x[2])

df_result['Current_Population'] = df_result['Current_Population'].astype(int)
df_result['Population_1970'] = df_result['Population_1970'].astype(int)

growth = df_result.apply(f, axis=1)         ### creates dictionary
df_growth = pd.DataFrame.from_dict(growth)  ### convert dictionary to dataframe
df_growth = df_growth.rename(columns={0: "52y Growth Rate"})
```

```python
df_more = df_result.join(df_growth)
#print(df_more)
```

```python
direction = 0
df_more = df_more.sort_values('52y Growth Rate', ascending=direction)
```
